## Introduction

The purpose of this notebook is to understand the Alternating Series Test:

**Theoerem (Alternating Series Test):** The alternating series $\sum_{k=1}^\infty (-1)^{k+1}a_k$ converges provided $0<a_{k+1}\le a_k$ and $\lim_{k\to\infty} a_k = 0$. 

**Remark** If the series is not alternating, then the above theorem is NOT true. Consider, for example $\sum_{k=1}^\infty \frac 1k$
